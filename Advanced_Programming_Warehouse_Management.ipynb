```python
#1 Define Inventory class

class Inventory:
    #define attributes of products
    def __init__(self, name: str="", quantity: int=0, price: int=0):
      self.name = name
      self.quantity = quantity
      self.price = price
```

```python
#2 Define Warehouse class

class Warehouse(Inventory):
  def __init__(self, name: str="", quantity: int = 0, price: float = 0) -> None:
    #call parent's init method
    super().__init__(name, quantity, price)
```

```python
#When initializing the program, prompt the user for action

def user_interface(iPhone, iPad, Mac):
    print("------- Hello and Welcome -------\n")
    print("We have the following products in our inventory: iPhones, iPads, and Macs")

#Create a loop so customer can buy again when the process is complete
    open = True
    while open:
        print("\n\n----- Inventory Management -----\n")
        print("Available options:")
        print("1 - View Inventory")
        print("2 - Add Items to Inventory")
        print("3 - Delete Items from Inventory")
        print("4 - Get the Quantity of a Particular Product")
        print("5 - Get the Price of a Particular Product")
        print("6 - Ask if Stock Available")
        print("7 - Exit\n")

#Prompt the user with which option to choose from and verify the input is an integer between 1 and 7
#Loop so the user gets prompt another time if wrong

        while True:
            try:
                option = int(input("Select an option between 1 and 7: "))
                if 1 <= option <= 7:
                    break
                raise ValueError()
            except ValueError:
                print("Please select an option between 1 and 7.")
    
#Loops over all possible decisions made by the user

#7 -  Exit
        if option == 7:
            print("Thank you for your visit!")
            open = False

#1 - Print report of all available items in the inventory
        if option == 1:
            print("The following items are available in stock:")
            print("There are ", iPhone.quantity, " iPhones in stock at the retail price of $", iPhone.price, ".", sep = "")
            print("There are ", iPad.quantity, " iPads in stock at the retail price of $", iPad.price, ".", sep = "")
            print("There are ", Mac.quantity, " Macs in stock at the retail price of $", Mac.price, ".", sep = "")

#2 - Add Items to Inventory
        elif option == 2:
            while True:
                try:
                    option_product = int(input("What would you like to add? Please choose one of the following options : 1 - iPhone, 2 - iPad, 3 - Mac, 0 - Go Back"))
                    
                    #iPhones
                    if option_product == 1:
                        while True:
                            try:
                                n_iPhone = int(input("How many iPhones would you like to add to inventory? Enter 0 to Go Back"))
                                assert(n_iPhone >= 0)
                                if n_iPhone == 0:
                                      break
                                iPhone.quantity = iPhone.quantity + n_iPhone
                                print(n_iPhone, "iPhones have been added from inventory.", iPhone.quantity, "iPhones are now in inventory.")
                            except:
                                print("Please enter a valid quantity")
                    
                    #iPads
                    elif option_product == 2:
                        while True:
                            try:
                                n_iPad = int(input("How many iPads would you like to add to inventory? Enter 0 to Go Back"))
                                assert(n_iPad >= 0)
                                if n_iPad == 0:
                                      break
                                iPad.quantity = iPad.quantity + n_iPad
                                print(n_iPad, "iPads have been added from inventory.", iPad.quantity, "iPads are now in inventory.")
                            except:
                                print("Please enter a valid quantity")
                      
                    #Macs
                    elif option_product == 3:
                        while True:
                            try:
                                n_Mac = int(input("How many Macs would you like to add to inventory? Enter 0 to Go Back"))
                                assert(n_Mac >= 0)
                                if n_Mac == 0:
                                    break
                                Mac.quantity = Mac.quantity + n_Mac
                                print(n_Mac, "Macs have been added from inventory.", Mac.quantity, "Macs hare now in inventory.")
                            except:
                                print("Please enter a valid quantity")

                    #Cancel
                    elif option_product == 0:
                      break

                except:
                    print("Please select a product among options 1-3 or Cancel")
                

#3 - Delete items from the inventory
        elif option == 3:
            while True:
                try:
                    option_product = int(input("What would you like to remove? Please input one of the following options : 1 - iPhone, 2 - iPad, 3 - Mac, 0 - Go Back"))
                    
                    #iPhones
                    if option_product == 1:
                        while True:
                            try:
                                n_iPhone = int(input("How many iPhones would you like to remove from inventory? Enter 0 to Go Back"))
                                assert(n_iPhone >= 0)
                                if n_iPhone == 0:
                                    break
                                iPhone.quantity = iPhone.quantity - n_iPhone
                                print(n_iPhone, "iPhones have been deleted from inventory.", iPhone.quantity, "iPhones are now in inventory.")
                            except:
                                print("Please enter a valid quantity")

                    #iPads
                    elif option_product == 2:
                        while True:
                            try:
                                n_iPad = int(input("How many iPads would you like to remove from inventory? Enter 0 to Go Back"))
                                assert(n_iPad >= 0)
                                if n_iPad == 0:
                                    break
                                iPad.quantity = iPad.quantity - n_iPad
                                print(n_iPad, "iPads have been deleted from inventory.", iPad.quantity, "iPads are now in inventory.")
                            except:
                                print("Please enter a valid quantity")

                    #Macs
                    elif option_product == 3:
                        while True:
                            try:
                                n_Mac = int(input("How many Macs would you like to remove from inventory? Enter 0 to Go Back"))
                                assert(n_Mac >= 0)
                                if n_Mac == 0:
                                    break
                                Mac.quantity = Mac.quantity - n_Mac
                                print(n_Mac, "Macs have been deleted from inventory.", Mac.quantity, "Macs are now in inventory.")
                            except:
                                print("Please enter a valid quantity")

                    #Cancel
                    elif option_product == 0:
                      break

                except:
                    print("Please select a product among options 1-3 or Cancel")


# 4 - Get the quantity from a particular item
        elif option == 4:
            while True:
                try:
                    option_product = int(input("What quantity would you like to know? Please input one of the following options : 1 - iPad, 2 - iPhone, 3 - Mac, 0 - Cancel"))
                    
                    #iPhones
                    if option_product == 1:
                        print(iPhone.quantity, "iPhones are currently in stock.")
                    
                    #iPads
                    if option_product == 2:
                        print(iPad.quantity, "iPads are currently in stock.")

                    #Macs
                    elif option_product == 3:
                        print(Mac.quantity, "Macs are currently in stock.")

                    #Cancel
                    elif option_product == 0:
                      break

                except:
                    print("Please select a product among options 1-3 or Cancel")
    

#5 - Get the price from a particular item
        elif option ==5:
            while True:
                try:
                    option_product = int(input("What price would you like to know? Please input one of the following options : 1 - iPad, 2 - iPhone, 3 - Mac, 0 - Cancel"))

                    #iPhones
                    if option_product == 1:
                        print("The price of an iPhone is $", iPhone.price, ".", sep = "")
                    
                    #iPads
                    elif option_product == 2:
                        print("The price of an iPad is $", iPad.price, ".", sep = "")
                                
                    #Macs
                    elif option_product == 3:
                        print("The price of a Mac is $",  Mac.price, ".", sep = "")

                    #Cancel
                    elif option_product == 0:
                      break

                except:
                    print("Please select a product among options 1-3 or Cancel")

#6 - Ask if a particular item is available
        elif option == 6:
            while True:
                try:
                    option_product = int(input("What item availability would you like to know? Please input one of the following options : 1 - iPhone, 2 - iPad, 3 - Mac, 0 - Cancel"))   

                    #iPads
                    if option_product == 1:
                        if iPhone.quantity > 0:
                                print("Yes, there are iPhones in stock!")
                        else:
                            print("Sorry, this product is currently unavailable...")
                    #iPhones
                    elif option_product ==2:
                        if iPad.quantity > 0:
                                print("Yes, there are iPads in stock!")
                        else:
                            print("Sorry, this product is currently unavailable...")
                    #Macs
                    elif option_product ==3:
                        if Mac.quantity > 0:
                            print("Yes, there are Macs in stock!")
                        else:
                            print("Sorry, this product is currently unavailable...")
                    
                    #Cancel
                    elif option_product == 0:
                      break

                except:
                    print("Please select a product among options 1-3 or Cancel")
```

```python
if __name__ == "__main__":
    iPhone = Warehouse("iPhone", 1000, 1000)
    iPad = Warehouse("iPad", 700, 1200)
    Mac = Warehouse("Mac", 500, 1500)
    user_interface(iPhone, iPad, Mac)
```
